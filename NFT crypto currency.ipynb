```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
rows = []
```

```python
def scrape():
    for i in range(1, 4):
        website = f"https://coinmarketcap.com/nft/collections/?page={i}"
        driver = webdriver.Chrome(r"C:\Users\menna\Desktop\chrome webdriver")
        driver.get(website)
        for j in range(1, 101):
            x_path = f'//*[@id="__next"]/div/div[1]/div[2]/div/div/div[3]/table/tbody/tr[{j}]'
            rows.append(driver.find_element('xpath', x_path).text)

scrape()
```

```python
len(rows)
```

```python
name = []
volume = []
market_cap = []
floor_price = []
Avg_price = []
sales = []
assets = []
owners = []
owners_per = []
for i in range(300):
    row = rows[i].split('\n')
    try:
        name.append(row[1])
        volume.append(row[2])
        market_cap.append(row[3])
        floor_price.append(row[4])
        Avg_price.append(row[5])
        sales.append(row[6])
        assets.append(row[7])
        owners.append(row[9])
        owners_per.append(row[8])
    except:
        pass

```

```python
cols = ['name','volume', 'market_cap', 'floor_price', 'Avg_price', 'sales', 'assets', 'owners', 'owners_per']
df = pd.DataFrame({
    'name':name,
    'volume':volume, 
    'market_cap':market_cap, 
    'floor_price':floor_price, 
    'Avg_price':Avg_price, 
    'sales':sales, 
    'assets':assets, 
    'owners':owners, 
    'owners_per': owners_per
})

df.to_csv('NFT_scraping')
```
